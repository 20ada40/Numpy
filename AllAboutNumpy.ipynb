STASTICAL FUNCTIONS

```python
a=np.array(5)
b=np.array([1,2,4,9])
c=np.array([[1,2,5,6],[6,7,3,4]])
d=np.array([[[2,3,4,5],[8,4,6,8]],[[4,7,6,7],[9,7,8,9]]])
print(np.amin(b))   #minimum value
print(np.amax(b))   # maximum value
print(np.amin(d))
print(np.amax(d))
print(np.mean(b))  #mean of the b array
print(np.mean(c))
e=np.array([1,3,5,87,9])
print(np.median(b))   #median of the b array
print(np.median(e))
print(np.std(b))   #standard deviation of the b array
print(np.std(c))
print(np.var(b))   #variance of the b array
print(np.var(c))
#average is similar to mean but the advantage of average is it also calculates the average weight of the graph
wght=np.array([2,4,6,8])
print(np.average(b))
print(np.average(b,weights=wght))
print(np.percentile(c,50)) #50 percent are less than 4  
print(np.percentile(c,50,axis=0))#axis=1 row, axis=0 is column
print(np.percentile(c,50,axis=1))
print(np.ptp(c)) #ptp is range
print(np.ptp(c,axis=1))
print(np.ptp(c,axis=0))
```
